```python
from hashlib import sha256

class User:
    def __init__(self, username, password, id, age=18):
        self.username = username    # public
        self.__password = password  # private - can be accessed throug method
        self.id = id
        self.age = age
    
    def get_info(self):
        print(f'username: {self.username}, id: {self.id}, age: {self.age}')
    
    def get_password_hash(self):
        hashed = sha256(self.__password.encode()).hexdigest()
        print(f'password hash: {hashed}')


p1 = User('marmag', '123', '00026')
p1.get_info()
p1.get_password_hash()
```

```python
import random

# losowa liczba całkowita z przedziału
print(f'Losowa liczba z przeciału 0-10: {random.randint(0, 10)}')

# losowa licza zmiennoprzecinkowa z przedziału 0-1
print(f'Losowa liczba z przeciału 0-1: {random.uniform(0, 1)}')

# wybór losowego elementu z listy
owoce = ['banan', 'jabłko', 'śliwka']
print(f'wybrano: {random.choice(owoce)}')

# losowa próbka z listy
liczby = [1, 2, 3, 4, 5]
print(f'próbka: {random.sample(liczby, 3)}')

# tasowanie listy
print(f'{liczby} -> ', end='')
random.shuffle(liczby)
print(f'{liczby}')

# kontrola losowości
random.seed(88)
```

```python
import pandas as pd
from pathlib import Path

# załadowanie pliku csv do DataFrame
# konwersja kolumny 'timestamp' na datetime
csv_path = Path('./example-files/logi.csv')
df = pd.read_csv(csv_path, parse_dates=['timestamp'])

# eksploracja danych
df.head()
# df.info()
# df.describe()

# operacje na kolumnach i filtorwanie
# df['Użytkownik']
# df[['Użytkownik', 'Status']]
# df[df['Status'] == 'FAILED']
# df['Użytkownik'].unique()
# df['Użytkownik'].value_counts()

df.groupby('Użytkownik').agg(
    wszystkie_próby = ('Status', 'count')
)

# praca z datami i godzinami
# pd.to_datetime(df['timestamp'])
# df['timestamp'].dt.date
# df['timestamp'].dt.hour

# sortowanie i czyszczenie danych

df.groupby('Użytkownik').agg(
    wszystkie_próby = ('Status', 'count')
    ).sort_values(by='wszystkie_próby', ascending=False)

# analiza zdarzeń w czasie

df.set_index('timestamp').resample('1min').size().to_dict()
```
